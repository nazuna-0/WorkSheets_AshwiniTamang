```python
# EXPERIMENTIN NUMPY VS SLOW PYTHON

import time
import numpy as np

size = 1_000_000
matrix_size = 1000

# Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

# Addition
start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Addition Time:", end - start)

# Multiplication
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Multiplication Time:", end - start)

# Dot Product
start = time.time()
dot = np.dot(arr1, arr2)
end = time.time()
print("NumPy Dot Product Time:", end - start)

# Matrix Multiplication
mat1 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
mat2 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
start = time.time()
mat_result = mat1 @ mat2
end = time.time()
print("NumPy Matrix Multiplication Time:", end - start)
```
